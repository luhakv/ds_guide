### 2. Области знаний в DS

#### Компьютерное зрение (CV)
Раздел DS, который специализируется на обработке изображений и видео. С инженерной точки зрения это способы лучше понять и сымитировать зрительный аппарат человека.

#### Обработка естественного языка (NLP)
Дисциплина, которая лежит на пересечении лингвистики, компьютерных наук и искусственного интеллекта. Основная задача - обработка/понимание естественного языка компьютерами.

#### Рекомендательные системы (RecSYS)
Дисциплина, которая ставит перед собой задачу предсказания, понравится человеку определенный объект или нет в определенный момент времени.

Временные последовательности (Time Series)
Дисциплина, которая направлена на выявление структуры временных рядов для их прогнозирования. Например, вы можете предсказать, стоит ли инкассации пополнять банкомат, учитывая историю его использования и множество других факторов.

#### Обработка аудио и речи (Audio&Speech Recognition)	
Дисциплина на пересечении компьютерных наук и лингвистики. Как в случае с NLP и CV, мы хотим научить компьютер обрабатывать аудиоданные схожим образом с тем, как это делают люди.

#### Машинное обучение на табличных данных (ML Tabular Data)
Всевозможные задачи классификации и регрессии на табличных данных. Обычно в таких задачах очень важную роль играют feature engineering и доменные знания.


### 3. Табличные данные: классические ML-задачи
**Табличные данные (tabular data)** — это, пожалуй, самый простой и понятный тип данных для машинного обучения. Имея в виду такой тип данных, вы буквально представляете таблицу. Каждая строка такой таблицы — это отдельный объект с перечнем своих признаков. В каждом столбце содержится по одному признаку, при этом один и тот же признак у разных объектов обязательно должен иметь один и тот же тип данных. Для работы с такими данными обычно используется библиотека Pandas.

Пример таких данных. Этот датасет называется "Churn's in Telecom Dataset" (https://bigml.com/user/francisco/gallery/dataset/5163ad540c0b5e5b22000383).

Обычно в таких данных все строки независимы, и вы работаете с каждой из них отдельно. 

У этого типа задач есть приятное свойство — они могут неплохо решаться эвристическими алгоритмами. 

Представьте, что вы хотите продать пользователю какой-то товар. Вы можете начать собирать данные, строить модели, измерять их эффективность и так далее, а можете просто придумать бизнес-метрику и правила (например, что если пользователь постоянно покупает у вас пиво, вы отправляете ему email с предложением купить сопутствующий товар со скидкой — чипсы или орешки). Такой подход гораздо проще: вы в короткий срок сможете оценить эффективность того или иного решения. 

✍️ Для сравнения подумайте, какую эвристику можно придумать для сервиса по распознаванию речи? Сами вы быстро построить такую систему не сможете, но в этих случаях можно использовать сторонние API.

Давайте посмотрим на примеры задач с табличным типом данных.

### ПРЕДСКАЗАНИЕ ОТТОКА КЛИЕНТОВ

Такую задачу обычно решают компании, в которых пользователи платят деньги ежемесячно, например операторы сотовой связи или сервисы по подписке. 

Компания хочет предугадать, что пользователь планирует покинуть сервис, и попытаться это предотвратить. Можно предложить более выгодный тариф или увеличить вовлечённость пользователя, например, рассказав про дополнительные возможности сервиса или предоставив на время какую-нибудь функциональность бесплатно. 

Собрать датасет для такой задачи обычно достаточно просто: у вас наверняка есть история того, как пользователь взаимодействовал с сервисом последний месяц. Вы собираете важные, на ваш взгляд, функцию, а целевую переменную вы можете определить как , если пользователь перестал платить, и как , если продолжил.

### ПРЕДСКАЗАНИЕ ЦЕНЫ ОБЪЕКТА

Представим, что у вас есть какой-то ценный объект, например дом, и вы хотите узнать цену, за которую этот дом можно продать. На платформе Kaggle компания Zillow проводила конкурс на предсказание цены дома (https://www.kaggle.com/c/zillow-prize-1). Победителю конкурса предлагался один миллион долларов — значит, задача достаточно востребованная! 

В наборе данных для такой задачи будут все возможные характеристики квартиры:

- этаж;
- площадь;
- год постройки;
- район;
- расстояние от метро;
- наличие магазинов;
- год капитального ремонта;
- и так далее. 

Целевой переменной окажется цена, за которую квартира была продана. Всю эту информацию достаточно просто собрать с сайтов продажи квартир. 

В современных подходах в таких задачах также используются фотографии — вы можете дополнить свои табличные данные информацией, которую можете вытащить из изображений, или даже делаете сложную модель, которая принимает на вход не только табличные данные, но и фотографии.

### ТЕХНОЛОГИИ

Обычно вам не нужны большие вычислительные мощности, чтобы построить модели для таких данных, поэтому вы можете попробовать сразу много моделей на своем датасете и выбрать лучшую. Можно экспериментировать с глубоким обучением и ансамблем моделей.

❗️ К слову о сложных моделях. Очень часто в таких задачах представители бизнеса или даже законы страны требуют интерпретируемости результатов вашей модели. Например, у вас есть модель, которая решает, выдавать ли клиенту банка кредит. В случае, если клиент обратится в суд, вам нужно будет объяснить, почему модель приняла именно такое решение, а не любое другое. В этом случае вы не можете использовать сложные модели. 

Обычно в таких задачах используют линейные модели, случайный лес, SVM, а также градиентный бустинг. В последнее время в таких задачах начали использовать Deep Learning. Эти методы могут отлично работать, но они плохо интерпретируемы. Основные библиотеки — это scikit-learn (линейные модели, случайный лес, SVM - https://scikit-learn.org/stable/), LightGBM (https://lightgbm.readthedocs.io/en/latest/) и CatBoost (https://catboost.ai/) для градиентного бустинга.

### 4. Табличные данные со своей спецификой
Сейчас мы поговорим о данных, которые с точки зрения представления или хранения не отличаются от табличных данных, которые мы уже обсудили. Это всё та же таблица, где строка является объектом, а столбцы — признаками. 

В чём же специфика таких данных? Давайте посмотрим на примере ранжирования поиска компании Spotify. 

### РАНЖИРОВАНИЕ ПОИСКА

Разработчики поиска хотят, чтобы пользователи как можно быстрее начинали слушать музыку и меньше времени проводили в поиске. Что для этого можно сделать? 

Допустим, пользователь ввел в поиске Юрий и хочет послушать песни авторства Юрия Хоя. Поисковая система достала из поискового индекса трёх исполнителей: Юрий Хой, Юрий Антонов и Юрий Шатунов. 

В каком порядке расположить этих исполнителей? Ближе к поисковой строке мы хотим расположить объекты, у которых вероятность быть выбранными выше. Как может выглядеть наш датасет? 

Сначала идут признаки пользователя, например вероятностное распределение по жанрам (если пользователь будет слушать трек, с какой вероятностью это будет рок, поп или академическая музыка?). 
Потом идут признаки артиста с теми же жанрами: если артист играет рок, то в этом признаке поставим , а если лёгкий рок, то  в признаке рок и  в признаке поп. 
Остальные признаки — это так называемое пересечение признаков артиста и пользователя. Искал ли пользователь этого артиста в последнее время? Добавил ли пользователь этого исполнителя в список любимых артистов? Вышел ли у артиста новый альбом? 
В этом случае мы хотим предсказать, кликнет пользователь на артиста из выдачи или нет. 

Чем такая задача отличается от предсказания кликнет/не кликнет? 

Разница заключается в том, что на выходе нам нужен ранжированный список объектов, а не предсказания по каждому объекту. Получается, в результате у нас есть пользователь, есть артист, мы подаём на вход все возможные фичи, связанные с этими объектами, затем добавляем фичи, которые одновременно описывают пользователя и артиста. 

В более сложных моделях мы можем добавить ещё один объект — текст запроса — и посчитать фичи, основанные на тексте запроса и имени артиста.

### РЕКОМЕНДАТЕЛЬНЫЕ СИСТЕМЫ

Похожий подход можно использовать в рекомендательных системах. Например, компания Netflix хочет вам прислать push-уведомление, чтобы вы обратили внимание на один из новых фильмов, который вышел на их платформе. 

Эта задача схожа с ранжированием. Можно взять все новые фильмы, данные по вашему профилю, посчитать фичи схожести и предсказать, у какого фильма большая вероятность быть просмотренным вами. Такой подход называется item to user — когда мы что-то знаем про пользователя и про объект и можем понять, что они подходят друг другу.

### КОЛЛАБОРАТИВНАЯ ФИЛЬТРАЦИЯ

Другой способ решить похожую задачу — коллаборативная фильтрация.

Логика этого подхода достаточно проста. Если пользователю понравились фильмы A, B и C, а второму пользователю фильмы B и C, то давайте посоветуем второму пользователю фильм A. 

Нам не надо знать никакие жанры и никакую специфику сервиса, просто достаточно помнить пользовательскую историю. В сервисе Netflix пользователи ставят оценки фильму — нравится или не нравится. Это и будут наши обучающие данные. 

В этом методе есть большая проблема — так называемая проблема холодного старта. Пока ни один пользователь не посмотрел фильм, мы не знаем, кому его стоит рекомендовать, а кому нет.

В современных подходах очень часто используют метод itemToVec. Это метод, который помогает нам перевести товар в векторное представление, и в этом векторном представлении товары, которые покупают вместе, будут близки друг к другу в плане косинусного расстояния. Такой подход используется в сервисах IKEA.

Вам не нужно осуществлять вход в сервисы IKEA, то есть вас не могут трекать как пользователя. К тому же, будем честны, покупки вы там делаете нечасто. В сервисе Spotify мы можем сказать, что если пользователь слушает рок, то давайте посоветуем рок, поп-рок, хард-рок и так далее. В случае с IKEA будет странно советовать пользователю диван, если он купил его год назад. Поэтому здесь для обучающей выборки используются корзины пользователей. 

Допустим, пользователь купил диван, стол и тумбу. Мы будем строить модель глубокого обучения таким образом: будем брать все объекты, кроме одного, и пробовать его. Например:

(диван, стол) → тумба

(диван, тумба) → стол

(стол, тумба) → диван

Таким образом, с помощью специальных моделей мы сможем получить векторное представление каждого купленного товара. Когда пользователь хочет купить какие-то вещи, мы можем порекомендовать ему что-то еще. Такой подход называется item to item: у нас нет знаний про пользователя, но мы можем предлагать рекомендации на основе товаров, которые он уже выбрал.

❗️ У этих трёх подходов есть свои плюсы и минусы, и выбор зависит от многих факторов. Усложнённое подобие последнего подхода было представлено компанией YouTube, и многие компании скопировали его, однако этот подход требует большого количества данных.

### РАСПОЗНАВАНИЕ ЛИЦ

У нас есть задача: на картинке или видеозаписи мы хотим найти все лица, а после этого идентифицировать каждого человека. Применений данной технологии может быть много:

→ Вы загрузили фотографию в социальную сеть, а мы хотим тут же отметить участников фотографии и отправить им уведомлению, чтобы увеличить вовлечённость пользователей.

→ Есть возможность предоставлять доступ по лицу. 

Компания Apple активно эксплуатирует эту возможность в своих телефонах, где вместо ввода кода или использования отпечатка пальца вы можете просто показать своё лицо. Говорят, что в новой версии iOS система научится распознавать людей даже в масках. 
С помощью такой фичи вы можете сделать доступ на предприятие, для чего на входе нужно поставить камеры и турникеты.
Недавно было анонсировано, что в Москве таким образом можно будет оплачивать проезд на метро.  

→ Слежка за людьми. Такая технология давно применяется в Китае: на улицах страны очень много камер видеонаблюдения. Если человек переходит дорогу на красный сигнал светофора, то система может распознать человека и впоследствии выписать ему штраф.

### СЕГМЕНТАЦИЯ

Иногда бывает нужным на картинке найти и выделить какой-то объект. Эта технология очень важна в медицинском компьютерном зрении. 

Например, на снимках мозга с магнитно-резонансной томографии необходимо распознать опухоль. У врачей не всегда хватает времени, чтобы детально рассмотреть снимки, у них также бывает разная квалификация, а ещё они, как многие люди, устают и не всегда сохраняют внимательность. 

Модели компьютерного зрения лишены многих из этих недостатков. При наличии достаточного количества хорошо размеченных данных они могут соревноваться с врачами. Современные медицинские модели работают не только с изображениями, но ещё и с данными пациентов, например: 

- рост;
- вес;
- пол;
- температура;
- место работы;
- предыдущей болезни;
- и так далее.

Они решают более сложные задачи — например, нужно ли делать операцию или опухоль безопасна. На сегодняшний день такие модели ещё работают в паре с врачом, а не самостоятельно.

Похожая задача решается в области беспилотных автомобилей. Как разработчики, мы хотим понимать, в какой среде находимся, а если рядом есть объект, то какого он класса. Вся эта информация подаётся на вход модели, которая определяет дальнейшие действия автомобиля.

Очень важное применение сегментации в условиях дистанционной работы в эпоху пандемии — это «заблюривание» или вставка фона во время видеозвонков. Эта функция пользуется большим спросом в продуктах Zoom и Hangouts и помогает скрыть беспорядок дома во время рабочих звонков :) Часто коллеги звонят «из космоса» или «с Бали», находясь в своих квартирах.

### ПОИСК КЛЮЧЕВЫХ ТОЧЕК

Возможно, вы помните, как некоторое время назад стали популярны различного рода маски в приложениях Instagram и Snapchat: пользователи могут надеть на себя различные маски или поменяться лицами. Эта функциональность была реализована с помощью разметки лица на специальные точки: если мы знаем положение лица в пространстве, то можем применить к нему спецэффекты.

❗️ По сравнению с задачами на табличных данных тут почти невозможно применить никаких эвристик. Допустим, у вас есть задача распознать, есть ли кот на картинке. Как программист, вы должны будете написать код, который по пикселям должен найти уши, усы, глаза, похожие на кошачьи, но это очень трудная задача. Ранние методы компьютерного зрения были основаны на поисках паттернов, но в последнее время они редко используются.

### 6. Обработка естественного языка (NLP)


На скриншоте вы видите англоязычную версию сайта РАН. Название Института белка́ был перевед как Squirrel Institute (институт белки). Непонятно, был сайт переведен человеком или компьютером, но такие ляпы можно встретить и сегодня при использовании популярных онлайн-переводчиков.

❗️ Основная цель обработки естественного языка схожа с компьютерным зрением: мы хотим, чтобы компьютеры научились понимать текст и делать с ним определённые действия так, как это делает человек.

### МАШИННЫЙ ПЕРЕВОД

Эта техника решает задачи перевода одной последовательности символов в другую, например перевод текста с одного языка на другой. Таких сервисов очень много: например, Google Translate или Яндекс.Переводчик — наверняка хотя бы раз в жизни вы пользовались подобным сервисом. В этой задаче у нас несколько параметров: язык ввода, текст и язык вывода.

В последнее время задача исправления опечаток решается похожим образом. На входе имеется текст с ошибками, а на выходе должен быть текст без ошибок. Датасет для такой задачи вы можете создать и сами, написав код, который превращает некорректное предложение в корректное, применяя к нему трансформацию. Вот только такой способ не будет до конца решать задачу, потому что ваша модель будет необъективна. Пример сервиса, который хорошо научился решать такую задачу — это Grammarly.

Также можно было наблюдать попытки создания сервисов, которые переводят естественный язык в SQL-запросы, например:

Средний возраст пользователей из Германии

↓

Select AVG(age) from users where country = 'germany'

Это своеобразная попытка компаний привлечь нетехнических людей к работе с данными, однако пока мало компаний достигло успехов в переводе естественного текста в SQL.

### ВОПРОСНО-ОТВЕТНЫЕ СИСТЕМЫ

Как правило, бизнес-идея таких систем состоит в том, чтобы снизить нагрузку на службу поддержки и дать им возможность вместо этого заниматься более сложными случаями. 

Во многих сервисах, если вы обращаетесь в службу поддержки онлайн, то первое время будете общаться с ботом. Современные боты достаточно неплохо справляются со своими задачами. Система обработки естественного языка пытается понять вашу проблему и предложить пути решения: часто это может быть ссылка на нужный раздел в «Часто задаваемых вопросах». Если же бот не может классифицировать ваш вопрос, то, скорее всего, к разговору присоединится живой человек и поможет решить ваш вопрос. 

Впоследствии такой разговор будет размечен, модель натренирована на новых данных, и, возможно, при следующем таком обращении бот уже сможет понять, что хочет пользователь. 

Обычно в таких системах роль NLP — это понимание вопроса и нахождения ответа в базе знаний. Например, мы спрашиваем: «Имя жены президента США». В этот момент система должна понять, что есть страна США, в которой есть президент, который женат, и у этой жены есть имя. Одно из последних интересных применений: модели скормили статьи про COVID-19 и сделали интерфейс, где любой человек может задать вопрос и получить ответ, основанный на статьях.

С недавнего времени появились более сложные боты, которые общаются с вами не онлайн, а звонят на телефон. Например, у компании IKEA есть бот, который может вам позвонить и договориться о более удобном времени доставки товара: узнает, есть ли у вас перед домом место для парковки грузовика, есть ли лифт и в какое время вы будете дома. В этой системе используется распознавание речи (перевод аудио в текст) и система перевода текста в аудио. Также вы можете задать боту вопросы, связанные с доставкой — на какие-то из них он сможет ответить.

### ТЕМАТИЧЕСКОЕ МОДЕЛИРОВАНИЕ

Тематическое моделирование — способ построения модели для коллекции текстовых документов, которая определяет, к каким темам относится каждый из документов. Такая модель может быть нужна в сервисе-агрегаторе новостей, где у вас каждый день появляется множество новых статей, и вы хотите их сортировать по специальным темам. Темы могут быть разного уровня, например:

- верхнеуровневые темы: политика, экономика, сельское хозяйство, спорт;
- более нишевые темы: футбол, бобслей, гонки;
- еще более нишевые темы: футбол в России, футбол в Англии.

Наличие такой классификации поможет вам создать удобный каталог для пользователей сайта, а также может послужить хорошими данными для моделей рекомендации.

Ещё более актуальна эта задача для сервисов с подкастами. Один из самых популярных подкастов на платформе Spotify — "The Joe Rogan Experience". Джо Роган выпускает подкаст почти каждый день, к нему приходят разные люди: режиссеры, бойцы UFC, музыканты, математики и люди, которые видели НЛО. Но идея в том, что не каждому интересны все темы, которые он освещает. По тексту подкаста мы можем найти отдельные части подкаста, найти для них тему и уже потом использовать эти знания при рекомендации или в поиске.

### ТЕХНОЛОГИИ

Хорошие новости заключаются в том, что для NLP не всегда нужны громадные датасеты и большие вычислительные мощности. Какое-то время назад дата-сайентисты пользовались методами классического машинного обучения, при этом они создавали фичи из самого текста. С появлением word to vec многие стали пользоваться векторами, например Glove или FastText, и стали строить модели поверх них. В последние год-два наступила эра трансформеров. Такие модели, как Bert, показали выдающиеся результаты на многих NLP-задачах, но вам понадобится много данных и вычислительных ресурсов, чтобы натренировать такие модели с нуля.

### 7. Последовательности со своей спецификой

Последний тип данных в нашей подборке — это последовательности со своей спецификой. В таких задачах текущее значение последовательности обычно зависит от предыдущих значений, поэтому нам трудно рассматривать текущий объект изолировано от предыдущих объектов. Давайте разбираться по порядку!

### РАСПОЗНАВАНИЕ РЕЧИ

Здесь мы коснёмся очень актуальной задачи перевода аудио в текст.

В последнее время появилось много умных колонок, основной интерфейс общения с которыми — голосовые команды. Обычно пайплайн работает так:

переводим аудио в текст с помощью модели распознавания речи

↓

классифицируем намерение с помощью NLP-модели

↓

выполняем запрос пользователя

Часто умные колонки оснащены двумя моделями распознавания речи. Одна — совсем простая, которая может реагировать на кодовые слова, например OK, Google. Такая модель работает прямо на «железе» самой колонки. Это сделано для того, чтобы сэкономить энергопотребление и отсылать меньше аудио на бэкенд нашего сервиса. После того, как колонка услышала кодовое слово, она начинает стримить аудио на бэкенд, где уже работает сложная система по распознавания речи.

Как говорилось ранее, такие системы могут использоваться для современных служб поддержки. Например, вы звоните в сервис по телемедицине, и до соединения с врачом система может опросить вас и узнать о текущем состоянии здоровья, чтобы врач начал рассматривать ваш случай, уже владея необходимой информацией.

Иногда такие системы используются в организациях, где нужна полная отчётность о взаимодействии пользователя с компанией. В некоторых странах банки обязаны хранить аудиоразговоры с клиентами и расшифровку этих разговоров в тексте. Раньше расшифровкой занимались операторы службы поддержки: они одновременно говорили с пользователем и печатали, но теперь им в этом помогают машины.

Для обучения с нуля хорошей системы распознавания речи обычно требуется около 10 тысяч часов данных. Под данными понимаются пары аудио → текст. По-хорошему нужно, чтобы данные были разнообразными (это применимо ко всем датасетам): там должны быть люди разных полов, с разными акцентами и разной скоростью речи. Тогда у вас получится хорошая модель.

Также при работе со звуком встречаются другие задачи. Например, на Kaggle проводилось соревнование: по аудио нужно было определить, присутствует ли на нём звук, который издает лягушка определенного вида.

### ПРЕДСКАЗАНИЕ НА ВРЕМЕННЫХ РЯДАХ

Временно́й ряд — собранный в разные моменты времени статистический материал о значении каких-либо параметров (в простейшем случае одного) исследуемого процесса. Иными словами, у нас есть значение какой-то переменной в разные периоды времени, и наша задача — предсказать значение этой переменной в будущем.

Давайте рассмотрим эту задачу на примере предсказания погоды: у нас есть много исторических данных о температуре, причём новые данные появляются каждый день. Нам не нужно ничего размечать, датасет полняется сам собой. 

Допустим, наша задача — предсказать температуру и осадки в городе Москва. Обычно люди интересуются, какая будет погода на следующий день. Возможные входные данные для нашей модели — это температура за сегодня и вчера... Но всё работает несколько сложнее: нам стоит добавить значения температуры вдали от Москвы за несколько сотен километров, и тогда из данных модель сможет понять, что на Москву движется теплый циклон со стороны Украины, поэтому, возможно, температура в Москве будет похожа на температуру в Харькове. 

❗️ То есть в таких ситуациях мы не просто используем исторические данные нашего города, но смотрим и на то, что происходит вокруг. Настоящие модели используют гораздо больше факторов, чем просто температура.

Похожая история с осадками: если мы знаем, что в какой-то части Московской области идёт дождь, то есть вероятность, что он может пойти и в какой-то части Москвы. Различия с температурой в том, что здесь нам нужны более свежие данные.

Интересный подход используется для прогнозирования погоды в местах, где данные с метеостанций задерживаются или их там просто нет. В таком случае используются снимки со спутников, и с помощью компьютерного зрения строится прогноз погоды.

### 9. Data Analyst, Data Engineer, Data Scientist

### DATA ANALYST

https://view.genial.ly/6048afde8fd9e60db9ea7b0d/interactive-content-data-analyst

Data Analyst — это специалист, который изучает, обрабатывает и истолковывает данные. 

Профессия лежит на стыке технологии и бизнеса — придется научиться находить общий язык с разными типами людей. Такая работа помогает лучше понимать пользователей и принимать решения при развитии продукта. Специалист должен обладать следующим набором навыков:

→ Технические:

уметь писать код, чтобы обрабатывать данные;
уметь работать с базами данных;
владеть инструментами для визуализации.
→ Презентационные: 

Очень часто при работе с бизнесом аналитику придется делать презентации, в которых нужно объяснять результаты A/B-тестов или истолковывать поведение продукта с учётом данных.

→ Математические: 

Знание статистики и методов машинного обучения.

### ТИПИЧНЫЕ ЗАДАЧИ

- Выгрузка и очистка данных;
- Exploratory Data Analysis;
- Анализ A/B-тестов;
- Создание и проведение презентаций;
- Анализ продукта.  

В этой роли ключевыми навыками являются не только технические, но и soft skills. С аналитиками работают, пожалуй, все разработчики, дата-инженеры, менеджеры, продукт-оунеры, руководители проектов, UX-дизайнеры и так далее.

Аналитики данных встречаются в основном в больших компаниях, в маленьких компаниях и стартапах такую роль может выполнять по чуть-чуть вся команда или технический продукт-оунер.

### DATA ENGINEER

https://view.genial.ly/6048c1c68fd9e60db9ea7d85/interactive-content-data-engineer

Data Engineer — программист по работе с данными. 

Это чисто техническая специальность. Лучший дата-инженер — это программист с хорошим опытом в сфере компьютерных наук. Он должен хорошо разбираться в распределённой обработке и хранении данных. Компании собирают о пользователях много данных, которые потом используются для исследования продукта, построения метрик, тренировки моделей.

### ТИПИЧНЫЕ ЗАДАЧИ

- Работа в команде с бэкенд-программистами для обеспечения сборки логов для нового продукта.
- Выгрузка данных из базы данных или распределенного хранения для дата-аналитика или дата-сайентиста.
- Написание пайплайна для обработки данных, который будет создавать датасет для последующего использования дата-сайентистом.
- Создание пайплайна, который будет считать ключевые метрики для продукта.
- Иногда — написание пайплайнов для тренировок моделей.  

Получается, что дата-инженеры — это программисты, которые специализируются на обработке данных. Обычно данных настолько много, что их невозможно обработать на одном компьютере, даже очень мощном, поэтому данные обрабатываются на многих компьютерах одновременно. Чтобы это делать, от дата-инженера требуется знание систем для параллельной обработки данных: Hadoop, Spark, Apache Beam и так далее.

Такие специалисты встречаются как в больших компаниях, так и в маленьких стартапах. Они в принципе могут быть совсем не вовлечены в бизнес-вопросы, но, как правило, чем выше уровень специалиста, тем больше он разбирается в бизнес-потребностях компании.

### DATA SCIENTIST

https://view.genial.ly/6048c3018fd9e60db9ea7dae/interactive-content-data-scientist

Если вы попросите несколько людей из этой области дать определение дата-сайентиста, то с большой вероятностью получите разные ответы. 

Можно сказать, что это разновидность программиста-математика, одной из главных задач которого является создание моделей машинного обучения. 

Как правило, в различных компаниях у людей с этой ролью очень разные компетенции. Про этот профиль можно сказать лишь то, что этот человек умеет делать fit/predict, а уже потом начинаются различия. 

В некоторых компаниях дата-сайнтисты занимаются только созданием и усовершенствованием моделей, то есть они не готовят данные (за них это делают дата-инженеры) и не выводят модели в продакшен (для этого есть бэкенд-разработчики). Такие специалисты в основном сосредоточены на тренировках моделей, feature-инжиниринге, улучшении метрик и поиске новых моделей. 

### ТИПИЧНЫЕ ЗАДАЧИ

- Тренировка моделей, оптимизация моделей, оптимизация обучения.
- Чтение статей, поиск новых решение.
- Поиск данных.
- Подготовка данных для тренировки(сбор логов, чистка данных).
- Подготовка данных для разметки, обучение разметчиков.

Реже:
- Вывод моделей в продакшен.
- Сбор метрик.

### 10. ML/DL Researcher

ML(DL) Researcher — это исследователь в области машинного обучения.

Главной задачей исследования является поиск новых знаний или систематическое расследование с целью установления фактов. Получается, что эти специалисты пробуют решать не решённые до этого момента времени задачи или улучшают старые решения. Чаще всего такие специалисты работают в университетах или больших компаниях. Для исследователей из компаний обычно стоит поиск новых решений в домене компаний.

Например, исследователи из компании Spotify первыми научились разделять вокал и музыку для аудиотреков. Такое умение нужно немногим, но для Spotify это позволило внедрить караоке-режим в приложении и улучшить распознавания текста в песнях. Иногда исследователи решают не совсем прикладные задачи: например, пока точно не понятно, как компании Alphabet пригодится умение бота DeepMind обыгрывать всех в StarCraft. 

Работа в компаниях и университетах различается. В первом случае у вас будет много вычислительных ресурсов, данных, финансирования для получения этих данных, а вот в университетах с этим похуже. Например, современный исследователь в NLP очень сильно зависит от вычислительных мощностей, и не всегда у исследователей из университета будет к ним доступ, в то время как, например, у Google нет проблем с доступом в дата-центры.

Если подытожить, то вам придется решать сложные задачи, которые пока никто не решал или решал, но вы сможете решить лучше. Обычно на эти роли берут людей похожих должностей из университетов.




### 11. ML/DL Engineer

ML/DL Engineer — специалист по полному циклу разработки моделей машинного обучения, начиная со сбора данных и заканчивая написанием бэкенд-сервисов. 

В этой специальности важен хороший уровень в Computer Science и знание алгоритмов машинного обучения.

### ТИПИЧНЫЕ ЗАДАЧИ

- Сбор данных.
- Анализ, очистка данных и создание датасетов для моделей.
- Поиск подходящих моделей для решения задачи.
- Тренировки моделей.
- Выкатка моделей в продакшен в бэкенд-сервисах/мобильных клиентах.
- Создание и сбор метрик.  

Такие специалисты часто «живут» в маленьких компаниях, где один работник может выполнять роли разных специалистов. ML/DL-инженер по сути должен выполнять роли дата-инженера и дата-сайентиста, а также иногда роль бэкенд-разработчика и очень редко — роль дата-аналитика. 

Иногда такие специалисты встречаются и в больших компаниях. С одной стороны, ваша команда может состоять из бэкендера, дата-инженера и дата-сайентиста, которые вместе делают бэкенд-сервис для модели машинного обучения. С другой стороны, такая же команда может состоять из трёх ML/DL-инженеров, которые владеют похожими компетенциями.

В то же время каждая из них предполагает глубокое погружение в узкоспециализированные вопросы, что даст вам экспертизу и высокую ценность на рынке труда.
