### 2. Области знаний в DS

#### Компьютерное зрение (CV)
Раздел DS, который специализируется на обработке изображений и видео. С инженерной точки зрения это способы лучше понять и сымитировать зрительный аппарат человека.

#### Обработка естественного языка (NLP)
Дисциплина, которая лежит на пересечении лингвистики, компьютерных наук и искусственного интеллекта. Основная задача - обработка/понимание естественного языка компьютерами.

#### Рекомендательные системы (RecSYS)
Дисциплина, которая ставит перед собой задачу предсказания, понравится человеку определенный объект или нет в определенный момент времени.

Временные последовательности (Time Series)
Дисциплина, которая направлена на выявление структуры временных рядов для их прогнозирования. Например, вы можете предсказать, стоит ли инкассации пополнять банкомат, учитывая историю его использования и множество других факторов.

#### Обработка аудио и речи (Audio&Speech Recognition)	
Дисциплина на пересечении компьютерных наук и лингвистики. Как в случае с NLP и CV, мы хотим научить компьютер обрабатывать аудиоданные схожим образом с тем, как это делают люди.

#### Машинное обучение на табличных данных (ML Tabular Data)
Всевозможные задачи классификации и регрессии на табличных данных. Обычно в таких задачах очень важную роль играют feature engineering и доменные знания.


### 3. Табличные данные: классические ML-задачи
**Табличные данные (tabular data)** — это, пожалуй, самый простой и понятный тип данных для машинного обучения. Имея в виду такой тип данных, вы буквально представляете таблицу. Каждая строка такой таблицы — это отдельный объект с перечнем своих признаков. В каждом столбце содержится по одному признаку, при этом один и тот же признак у разных объектов обязательно должен иметь один и тот же тип данных. Для работы с такими данными обычно используется библиотека Pandas.

Пример таких данных. Этот датасет называется "Churn's in Telecom Dataset" (https://bigml.com/user/francisco/gallery/dataset/5163ad540c0b5e5b22000383).

Обычно в таких данных все строки независимы, и вы работаете с каждой из них отдельно. 

У этого типа задач есть приятное свойство — они могут неплохо решаться эвристическими алгоритмами. 

Представьте, что вы хотите продать пользователю какой-то товар. Вы можете начать собирать данные, строить модели, измерять их эффективность и так далее, а можете просто придумать бизнес-метрику и правила (например, что если пользователь постоянно покупает у вас пиво, вы отправляете ему email с предложением купить сопутствующий товар со скидкой — чипсы или орешки). Такой подход гораздо проще: вы в короткий срок сможете оценить эффективность того или иного решения. 

✍️ Для сравнения подумайте, какую эвристику можно придумать для сервиса по распознаванию речи? Сами вы быстро построить такую систему не сможете, но в этих случаях можно использовать сторонние API.

Давайте посмотрим на примеры задач с табличным типом данных.

### ПРЕДСКАЗАНИЕ ОТТОКА КЛИЕНТОВ

Такую задачу обычно решают компании, в которых пользователи платят деньги ежемесячно, например операторы сотовой связи или сервисы по подписке. 

Компания хочет предугадать, что пользователь планирует покинуть сервис, и попытаться это предотвратить. Можно предложить более выгодный тариф или увеличить вовлечённость пользователя, например, рассказав про дополнительные возможности сервиса или предоставив на время какую-нибудь функциональность бесплатно. 

Собрать датасет для такой задачи обычно достаточно просто: у вас наверняка есть история того, как пользователь взаимодействовал с сервисом последний месяц. Вы собираете важные, на ваш взгляд, функцию, а целевую переменную вы можете определить как , если пользователь перестал платить, и как , если продолжил.

### ПРЕДСКАЗАНИЕ ЦЕНЫ ОБЪЕКТА

Представим, что у вас есть какой-то ценный объект, например дом, и вы хотите узнать цену, за которую этот дом можно продать. На платформе Kaggle компания Zillow проводила конкурс на предсказание цены дома (https://www.kaggle.com/c/zillow-prize-1). Победителю конкурса предлагался один миллион долларов — значит, задача достаточно востребованная! 

В наборе данных для такой задачи будут все возможные характеристики квартиры:

- этаж;
- площадь;
- год постройки;
- район;
- расстояние от метро;
- наличие магазинов;
- год капитального ремонта;
- и так далее. 

Целевой переменной окажется цена, за которую квартира была продана. Всю эту информацию достаточно просто собрать с сайтов продажи квартир. 

В современных подходах в таких задачах также используются фотографии — вы можете дополнить свои табличные данные информацией, которую можете вытащить из изображений, или даже делаете сложную модель, которая принимает на вход не только табличные данные, но и фотографии.

### ТЕХНОЛОГИИ

Обычно вам не нужны большие вычислительные мощности, чтобы построить модели для таких данных, поэтому вы можете попробовать сразу много моделей на своем датасете и выбрать лучшую. Можно экспериментировать с глубоким обучением и ансамблем моделей.

❗️ К слову о сложных моделях. Очень часто в таких задачах представители бизнеса или даже законы страны требуют интерпретируемости результатов вашей модели. Например, у вас есть модель, которая решает, выдавать ли клиенту банка кредит. В случае, если клиент обратится в суд, вам нужно будет объяснить, почему модель приняла именно такое решение, а не любое другое. В этом случае вы не можете использовать сложные модели. 

Обычно в таких задачах используют линейные модели, случайный лес, SVM, а также градиентный бустинг. В последнее время в таких задачах начали использовать Deep Learning. Эти методы могут отлично работать, но они плохо интерпретируемы. Основные библиотеки — это scikit-learn (линейные модели, случайный лес, SVM - https://scikit-learn.org/stable/), LightGBM (https://lightgbm.readthedocs.io/en/latest/) и CatBoost (https://catboost.ai/) для градиентного бустинга.

### 4. Табличные данные со своей спецификой
Сейчас мы поговорим о данных, которые с точки зрения представления или хранения не отличаются от табличных данных, которые мы уже обсудили. Это всё та же таблица, где строка является объектом, а столбцы — признаками. 

В чём же специфика таких данных? Давайте посмотрим на примере ранжирования поиска компании Spotify. 

### РАНЖИРОВАНИЕ ПОИСКА

Разработчики поиска хотят, чтобы пользователи как можно быстрее начинали слушать музыку и меньше времени проводили в поиске. Что для этого можно сделать? 

Допустим, пользователь ввел в поиске Юрий и хочет послушать песни авторства Юрия Хоя. Поисковая система достала из поискового индекса трёх исполнителей: Юрий Хой, Юрий Антонов и Юрий Шатунов. 

В каком порядке расположить этих исполнителей? Ближе к поисковой строке мы хотим расположить объекты, у которых вероятность быть выбранными выше. Как может выглядеть наш датасет? 

Сначала идут признаки пользователя, например вероятностное распределение по жанрам (если пользователь будет слушать трек, с какой вероятностью это будет рок, поп или академическая музыка?). 
Потом идут признаки артиста с теми же жанрами: если артист играет рок, то в этом признаке поставим , а если лёгкий рок, то  в признаке рок и  в признаке поп. 
Остальные признаки — это так называемое пересечение признаков артиста и пользователя. Искал ли пользователь этого артиста в последнее время? Добавил ли пользователь этого исполнителя в список любимых артистов? Вышел ли у артиста новый альбом? 
В этом случае мы хотим предсказать, кликнет пользователь на артиста из выдачи или нет. 

Чем такая задача отличается от предсказания кликнет/не кликнет? 

Разница заключается в том, что на выходе нам нужен ранжированный список объектов, а не предсказания по каждому объекту. Получается, в результате у нас есть пользователь, есть артист, мы подаём на вход все возможные фичи, связанные с этими объектами, затем добавляем фичи, которые одновременно описывают пользователя и артиста. 

В более сложных моделях мы можем добавить ещё один объект — текст запроса — и посчитать фичи, основанные на тексте запроса и имени артиста.

### РЕКОМЕНДАТЕЛЬНЫЕ СИСТЕМЫ

Похожий подход можно использовать в рекомендательных системах. Например, компания Netflix хочет вам прислать push-уведомление, чтобы вы обратили внимание на один из новых фильмов, который вышел на их платформе. 

Эта задача схожа с ранжированием. Можно взять все новые фильмы, данные по вашему профилю, посчитать фичи схожести и предсказать, у какого фильма большая вероятность быть просмотренным вами. Такой подход называется item to user — когда мы что-то знаем про пользователя и про объект и можем понять, что они подходят друг другу.

### КОЛЛАБОРАТИВНАЯ ФИЛЬТРАЦИЯ

Другой способ решить похожую задачу — коллаборативная фильтрация.

Логика этого подхода достаточно проста. Если пользователю понравились фильмы A, B и C, а второму пользователю фильмы B и C, то давайте посоветуем второму пользователю фильм A. 

Нам не надо знать никакие жанры и никакую специфику сервиса, просто достаточно помнить пользовательскую историю. В сервисе Netflix пользователи ставят оценки фильму — нравится или не нравится. Это и будут наши обучающие данные. 

В этом методе есть большая проблема — так называемая проблема холодного старта. Пока ни один пользователь не посмотрел фильм, мы не знаем, кому его стоит рекомендовать, а кому нет.

В современных подходах очень часто используют метод itemToVec. Это метод, который помогает нам перевести товар в векторное представление, и в этом векторном представлении товары, которые покупают вместе, будут близки друг к другу в плане косинусного расстояния. Такой подход используется в сервисах IKEA.

Вам не нужно осуществлять вход в сервисы IKEA, то есть вас не могут трекать как пользователя. К тому же, будем честны, покупки вы там делаете нечасто. В сервисе Spotify мы можем сказать, что если пользователь слушает рок, то давайте посоветуем рок, поп-рок, хард-рок и так далее. В случае с IKEA будет странно советовать пользователю диван, если он купил его год назад. Поэтому здесь для обучающей выборки используются корзины пользователей. 

Допустим, пользователь купил диван, стол и тумбу. Мы будем строить модель глубокого обучения таким образом: будем брать все объекты, кроме одного, и пробовать его. Например:

(диван, стол) → тумба

(диван, тумба) → стол

(стол, тумба) → диван

Таким образом, с помощью специальных моделей мы сможем получить векторное представление каждого купленного товара. Когда пользователь хочет купить какие-то вещи, мы можем порекомендовать ему что-то еще. Такой подход называется item to item: у нас нет знаний про пользователя, но мы можем предлагать рекомендации на основе товаров, которые он уже выбрал.

❗️ У этих трёх подходов есть свои плюсы и минусы, и выбор зависит от многих факторов. Усложнённое подобие последнего подхода было представлено компанией YouTube, и многие компании скопировали его, однако этот подход требует большого количества данных.

### РАСПОЗНАВАНИЕ ЛИЦ

У нас есть задача: на картинке или видеозаписи мы хотим найти все лица, а после этого идентифицировать каждого человека. Применений данной технологии может быть много:

→ Вы загрузили фотографию в социальную сеть, а мы хотим тут же отметить участников фотографии и отправить им уведомлению, чтобы увеличить вовлечённость пользователей.

→ Есть возможность предоставлять доступ по лицу. 

Компания Apple активно эксплуатирует эту возможность в своих телефонах, где вместо ввода кода или использования отпечатка пальца вы можете просто показать своё лицо. Говорят, что в новой версии iOS система научится распознавать людей даже в масках. 
С помощью такой фичи вы можете сделать доступ на предприятие, для чего на входе нужно поставить камеры и турникеты.
Недавно было анонсировано, что в Москве таким образом можно будет оплачивать проезд на метро.  

→ Слежка за людьми. Такая технология давно применяется в Китае: на улицах страны очень много камер видеонаблюдения. Если человек переходит дорогу на красный сигнал светофора, то система может распознать человека и впоследствии выписать ему штраф.

### СЕГМЕНТАЦИЯ

Иногда бывает нужным на картинке найти и выделить какой-то объект. Эта технология очень важна в медицинском компьютерном зрении. 

Например, на снимках мозга с магнитно-резонансной томографии необходимо распознать опухоль. У врачей не всегда хватает времени, чтобы детально рассмотреть снимки, у них также бывает разная квалификация, а ещё они, как многие люди, устают и не всегда сохраняют внимательность. 

Модели компьютерного зрения лишены многих из этих недостатков. При наличии достаточного количества хорошо размеченных данных они могут соревноваться с врачами. Современные медицинские модели работают не только с изображениями, но ещё и с данными пациентов, например: 

- рост;
- вес;
- пол;
- температура;
- место работы;
- предыдущей болезни;
- и так далее.

Они решают более сложные задачи — например, нужно ли делать операцию или опухоль безопасна. На сегодняшний день такие модели ещё работают в паре с врачом, а не самостоятельно.

Похожая задача решается в области беспилотных автомобилей. Как разработчики, мы хотим понимать, в какой среде находимся, а если рядом есть объект, то какого он класса. Вся эта информация подаётся на вход модели, которая определяет дальнейшие действия автомобиля.

Очень важное применение сегментации в условиях дистанционной работы в эпоху пандемии — это «заблюривание» или вставка фона во время видеозвонков. Эта функция пользуется большим спросом в продуктах Zoom и Hangouts и помогает скрыть беспорядок дома во время рабочих звонков :) Часто коллеги звонят «из космоса» или «с Бали», находясь в своих квартирах.

### ПОИСК КЛЮЧЕВЫХ ТОЧЕК

Возможно, вы помните, как некоторое время назад стали популярны различного рода маски в приложениях Instagram и Snapchat: пользователи могут надеть на себя различные маски или поменяться лицами. Эта функциональность была реализована с помощью разметки лица на специальные точки: если мы знаем положение лица в пространстве, то можем применить к нему спецэффекты.

### ТЕХНОЛОГИИ

Обычно такие задачи решаются с помощью моделей глубокого обучения, а точнее, с помощью свёрточных сетей. Задачи требуют больших вычислительных ресурсов и большого количества тренировочных данных. Для построения моделей используют библиотеки Tensorflow, Keras (часть Tensorflow), PyTorch, с которыми вы уже знакомы. 

Для многих задач существуют предобученные модели: если вы работаете в маленькой компании, то вы можете использовать их или доучить на ваших данных, а уже затем использовать.
